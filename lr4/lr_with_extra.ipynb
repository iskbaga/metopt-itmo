```python
import numpy as np
from scipy.optimize import minimize
import matplotlib.pyplot as plt
```

```python
class f_rosenbrock():

  def f(x):
    return ((1 - x[0])**2) + (100 * (x[1] - (x[0]**2))**2)

  def grad(x):
    return np.array([
        -2 * (1 - x[0]) - 4 * 100 * (x[1] - x[0]**2) * x[0],
        2 * 100 * (x[1] - x[0]**2)
    ])

  def hess(x):
    return np.matrix([
        [1200 * x[0]**2 - 400 * x[1] + 2, -400 * x[0]],
        [-400 * x[0], 200]
    ])

  def domain():
    return (-5, 5)

class f_himmelblau():

  def f(x):
    return (x[0]**2 + x[1] - 11)**2 + (x[0] + x[1]**2 - 7)**2

  def grad(x):
    return np.array([
          2 * (x[0]**2 + x[1] - 11) * (2 * x[0]) + 2 * (x[0] + x[1]**2 - 7),
          2 * (x[0]**2 + x[1] - 11) + 2 * (x[0] + x[1] **2 - 7) * (2 * x[1])
      ])

  def hess(x):
    return np.matrix([
      [8 * x[0]**2 + 4 * (x[0]**2 + x[1] - 11) + 2, 4 * x[0] + 4 * x[1]],
      [4 * x[0] + 4 * x[1], 8 * x[1]**2 + 4 * (x[1]**2 + x[0] - 7) + 2]
    ])

  def domain():
    return (-5, 5)

class f_rastrigin():

  def f(x):
    return 20 + (x[0]**2 - 10 * np.cos(2 * np.pi * x[0])) + (x[1]**2 - 10 * np.cos(2 * np.pi * x[1]))

  def grad(x):
    return np.array([
          2 * x[0] + 20 * np.pi * np.sin(2 * np.pi * x[0]),
          2 * x[1] + 20 * np.pi * np.sin(2 * np.pi * x[1])
      ])

  def hess(x):
    return np.matrix([
      [40 * np.pi**2 * np.cos(2 * np.pi * x[0]) + 2, 0],
      [0, 40 * np.pi**2 * np.cos(2 * np.pi * x[1]) + 2]
    ])

  def domain():
    return (-5, 10)
```

```python
def grad_step(lr_func):

  def grad_inner(f, x = None, maxiter = 1000, tol = 1e-18):
    if x is None:
        x = np.random.random(2)

    for iter in range(1, maxiter+1):
        grad = f.grad(x)
        lr = lr_func(f.f, x, -grad)
        if np.linalg.norm(lr * grad) < tol:
            return x, iter
        x -= lr * grad

    return x, maxiter

  return grad_inner

def newton(lr_func):

  def newton_inner(f, x = None, maxiter = 1000, tol = 1e-18):
    if x is None:
        x = np.random.random(2)

    prev_x = np.copy(x)
    for iter in range(1, maxiter+1):
      hess = f.hess(x)
      grad = f.grad(x)
      p = np.linalg.solve(hess, -grad)
      lr = lr_func(f.f, x, p)
      x += lr * p

      if np.linalg.norm(x - prev_x) < tol:
        return x, iter
      prev_x = np.copy(x)

    return x, maxiter

  return newton_inner

def fixed_search(f, x, d, tol = None):
    return 1

def dichotomy_search(f, x, d, tol=1e-6):
    a, b = -1, 1

    while abs(b - a) > tol:
        c = (a + b) / 2
        x1, x2 = x + (c - tol) * d, x + (c + tol) * d
        f1, f2 = f(x1), f(x2)

        if f1 < f2:
            b = c
        else:
            a = c

    return (a + b) / 2

def golden_section_search(f, x, d, tol=1e-6):
    a, b = 0, 1
    rho = (np.sqrt(5) - 1) / 2

    while abs(b - a) > tol:
        x1 = a + (1 - rho) * (b - a)
        x2 = a + rho * (b - a)

        f1 = f(x + x1 * d)
        f2 = f(x + x2 * d)

        if f1 < f2:
            b = x2
        else:
            a = x1

    return (a + b) / 2

def Nelder_Mead_search(f, curr_x, num_iter = 1000):
    history = []
    def callback_func(x):
        history.append(np.copy(x))
    res = minimize(f, curr_x, method='Nelder-Mead', options={'disp': True, 'maxiter' : num_iter}, callback=callback_func)
    return res.nit, history, res.x

def minimize_wrapper(method, options = {}, jac = None):

  def inner(f, x0 = None, maxiter = 1000, tol = 1e-18):
    if x0 is None:
      x0 = np.random.random(2)

    options['maxiter'] = maxiter

    _jac = jac
    if _jac is None:
      _jac = f.grad
    res = minimize(f.f, x0, method=method, jac=_jac, hess=f.hess, tol=tol, options=options)
    return res.x, res.nit

  return inner
```

```python
class method_class:
  def __init__(self, name, f):
      self.name = name
      self.f = f

newton_fixed = method_class("Ньютон с фиксированным шагом", newton(fixed_search))
newton_dichotomy = method_class("Ньютон с дихотомией", newton(dichotomy_search))
newton_cg = method_class("Newton-CG", minimize_wrapper('Newton-CG'))

nelder_mead = method_class("Neldear-Mead", minimize_wrapper('Nelder-Mead'))
grad_golden = method_class("Градиентный спуск с золотым сечением", grad_step(golden_section_search))

bfgs = method_class("BFGS", minimize_wrapper('BFGS', options = {'norm': 2}))
lbfgs = method_class("L-BFGS-B", minimize_wrapper('L-BFGS-B'))
bfgs_diff = method_class("BFGS, производная через разностые схемы", minimize_wrapper('BFGS', options = {'norm': 2}, jac = '3 step'))
lbfgs_diff = method_class("L-BFGS-B, производная через разностые схемы", minimize_wrapper('L-BFGS-B', jac = '3 step'))
```

```python
class method_result:

  def __init__(self):
    self.history = []
    self.fev = 0
    self.jacev = 0
    self.hessev = 0

class f_wrapper:

  def __init__(self, inner_f):
    self.inner_f = inner_f
    self.result = method_result()

  def record(self, x):
    self.result.history.append(np.around(x, 5))

  def f(self, x):
    self.record(x)
    self.result.fev += 1
    return self.inner_f.f(x)

  def grad(self, x):
    self.record(x)
    self.result.jacev += 1
    return self.inner_f.grad(x)

  def hess(self, x):
    self.record(x)
    self.result.hessev += 1
    return self.inner_f.hess(x)

def record_calc(inner_f, to_run):
  f = f_wrapper(inner_f)
  return to_run(f), f.result

def analyze(f, x0, method):
  return record_calc(f, lambda fu: method(fu, x0))

# record_calc(f_rosenbrock, lambda f: newton(f, fixed_search))
# record_calc(f_rosenbrock, lambda f: newton(f, dichotomy_search))
# newton_cg(f_rosenbrock)
# res, his = record_calc(f_rosenbrock, lambda f: newton_cg(f))
# print(res)
# print(len(his))
# analyze(f_spherical, None, newton(fixed_search))
```

```python
import random

def decode_2ints(xy):
  x = xy >> 32
  y = xy - (x << 32)
  return (x, y)

def chromosome_decoding(x, cube):
  a, b = cube
  xk, yk = decode_2ints(x)
  len_coef = (b - a) / (1 << 32)
  return [a + xk * len_coef, a + yk * len_coef]

chromosome_decoding(random.getrandbits(64), (2, 4))

class chromosome:
  def __init__(self, f, bin_x):
      x = chromosome_decoding(bin_x, f.domain())
      self.x = x
      self.bin_x = bin_x
      self.val = -f.f(x)

def sort_by_val(population):
  population.sort(key = lambda chromo: chromo.val, reverse = True)

def form_population(f, xs):
  population = list(map(lambda x: chromosome(f, x), xs))
  sort_by_val(population)
  return population

def tournament_selection(population, cnt, k = 10):
  ans = []
  for _ in range(cnt):
    sample = random.sample(population, k)
    winner = max(sample, key = lambda chromo: chromo.val) # we can just take the smallest index instead
    ans.append(winner.bin_x)
  return ans

def divide(x, point):
  left = x >> point << point
  right = x - left
  return (left, right)

def one_point_crossover(a, b, crossover_prob = 0.5):
  if random.random() > crossover_prob:
    return (a, b)
  point = random.randint(1, 63)
  a_left, a_right = divide(a, point)
  b_left, b_right = divide(b, point)
  return (a_left + b_right, a_right + b_left)

def mutate_and(x, degree = 4):
  mask = random.getrandbits(64)
  for i in range(degree - 1):
    mask &= random.getrandbits(64)
  return x ^ mask

# (population_size - elite_cnt) % 2 == 0
def genetic(f, population_size, elite_cnt, selection, crossover, mutate, maxiter=int(1e18), tol=1e-9, tries=20):
  initial = [random.getrandbits(64) for _ in range(population_size)]
  cur_population = form_population(f, initial)

  stop_cnt = 0
  for i in range(maxiter):
    new_x = [chromo.bin_x for chromo in cur_population[:elite_cnt]]

    parents = selection(cur_population, population_size - elite_cnt)
    offspring = [crossover(parents[j], parents[j+1]) for j in range(0, len(parents), 2)]
    offspring = [item for t in offspring for item in t] # list of pairs to list
    new_x.extend(offspring)

    new_x = list(map(mutate, new_x))
    new_population = form_population(f, new_x)
    if new_population[0].val - cur_population[0].val < tol:
      stop_cnt += 1
      if stop_cnt == tries:
        break
    else:
      stop_cnt = 0
    cur_population = new_population

    cur_x = np.array([np.array(chromo.x) for chromo in cur_population])

    if i % 10 == 0:
      x = np.linspace(-6, 6, 100)
      y = np.linspace(-6, 6, 100)
      X, Y = np.meshgrid(x, y)
      Z = f.f(np.array([X, Y]))
      fig, ax = plt.subplots(figsize=(8, 6))
      ax.contour(X, Y, Z, levels=np.logspace(0, 5, 35), cmap='jet')
      ax.scatter(cur_x[:, 0], cur_x[:, 1], color='red')
      ax.set_xlabel('X')
      ax.set_ylabel('Y')
      plt.show()

  return chromosome_decoding(cur_population[0].bin_x, f.domain())

genetic(f_himmelblau, 1000, 10, tournament_selection, one_point_crossover, mutate_and)
```

```python
def newton_optuna_objective(trial, function):

    maxiter = trial.suggest_int('maxiter', 100, 1000)
    tol = trial.suggest_float('tol', 1e-12, 1e-6, log=True)

    def newton(f, x = None, maxiter = maxiter, tol = tol):
        if x is None:
            x = np.random.random(2)

        prev_x = np.copy(x)
        for iter in range(1, maxiter+1):
            hess = f.hess(x)
            grad = f.grad(x)
            p = np.linalg.solve(hess, -grad)
            x += p

            if np.linalg.norm(x - prev_x) < tol:
                return x, iter
            prev_x = np.copy(x)

        return x, maxiter

    x0 = np.random.random(2)
    _, n_iter = newton(function, x0)
    return n_iter

def optimize_function(func_class, n_trials=50):
    study = optuna.create_study(direction='minimize')
    study.optimize(lambda trial: newton_optuna_objective(trial, func_class), n_trials=n_trials)

    print(f"Results for {func_class.__class__.__name__}:")
    print("Number of trials: ", len(study.trials))
    print("Best trial:")
    trial = study.best_trial
    print("  Value: ", trial.value)
    print("  Params: ")
    for key, value in trial.params.items():
        print(f"    {key}: {value}")
    return study

study1 = optimize_function(f_rosenbrock, n_trials=50)
study2 = optimize_function(f_himmelblau, n_trials=50)
study3 = optimize_function(f_rastrigin, n_trials=50)

```
